```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('PARTICIPACAO-EM-CONGRESSO')



#Parte A
natureza = []
titulo = []
ano = []
tipo_participacao = []
forma_participacao = []

#Parte B

nome_evento = []


for i in b_unique:

    a = i.find("DADOS-BASICOS-DA-PARTICIPACAO-EM-CONGRESSO")
   
    
    natureza.append(a.get('NATUREZA'))
    titulo.append(a.get('TITULO'))
    ano.append(a.get("ANO"))
    tipo_participacao.append(a.get('TIPO-PARTICIPACAO'))
    forma_participacao.append(a.get('FORMA-PARTICIPACAO'))
    
    b = i.find("DETALHAMENTO-DA-PARTICIPACAO-EM-CONGRESSO")
    nome_evento.append(b.get('NOME-DO-EVENTO'))

              
df = pd.DataFrame()
df['titulo'] = titulo
df['natureza'] = natureza
df['ano'] = ano
df['tipo_participacao'] = tipo_participacao
df['forma_participacao'] = forma_participacao
df['nome_evento'] = nome_evento



        
    

```

```python
df = df.sort_values(by=['ano'], ascending=False)
```
