```python
from bs4 import BeautifulSoup
from googletrans import Translator
import pandas as pd
```

```python
# Reading data from the xml file
with open('cv3.xml', 'r', encoding='utf-8') as f:
    data = f.read()
 
```

```python
translator = Translator()
```

```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('ARTIGO-PUBLICADO')


total = 0
primeiro = 0
supervisor = 0

titulo = []
doi = []
ano = []
home_page = []
periodico = []
autores = []

for i in b_unique:
    total = total+1
    a = i.find("DADOS-BASICOS-DO-ARTIGO")
   
    b = i.find("DETALHAMENTO-DO-ARTIGO")
    

    
    titulo.append(a.get('TITULO-DO-ARTIGO').title())
    doi.append(a.get('DOI'))
    ano.append(a.get("ANO-DO-ARTIGO"))
    home_page.append(a.get('HOME-PAGE-DO-TRABALHO'))
    
    periodico.append(b.get("TITULO-DO-PERIODICO-OU-REVISTA").title())
    r = i.find_all('AUTORES')
    
  
        
    n_autores = 0
    s_autores = ''
    for k in r:
        s_autores = s_autores+ k.get('NOME-PARA-CITACAO').title()
        n_autores = n_autores+1
        if('lobato' in k.get('NOME-PARA-CITACAO').lower()):
           
            if(autores ==1):
                primeiro = primeiro+1

            if(autores == len(r)):
                supervisor = supervisor+1
            

    autores.append(s_autores)
print(primeiro, supervisor, total)                  
df = pd.DataFrame()
df['titulo'] = titulo
df['doi'] = doi
df['ano'] = ano
df['home_page'] = home_page
df['periodico'] = periodico
df['autores'] = autores

df.head(13)

        
    

```
