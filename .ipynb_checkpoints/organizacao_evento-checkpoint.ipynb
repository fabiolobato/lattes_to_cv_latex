```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('SOFTWARE')


titulo = []
ano = []
codigo_registro = []
autores = []


total = 0
primeiro = 0
supervisor = 0

for i in b_unique:

    a = i.find("DADOS-BASICOS-DO-SOFTWARE")
   
    b = i.find("DETALHAMENTO-DO-SOFTWARE")
    
    c = b.find('REGISTRO-OU-PATENTE')
    if(c!=None):
        
        total = total+1
    
    
        r = i.find_all('AUTORES')



        titulo.append(a.get('TITULO-DO-SOFTWARE'))
        ano.append(a.get("ANO"))
        codigo_registro.append(c.get('CODIGO-DO-REGISTRO-OU-PATENTE'))
        
        n_autores = 0
        s_autores = ''
        for k in r:
            s_autores = s_autores+ k.get('NOME-PARA-CITACAO').title()+'; '
            n_autores = n_autores+1
            if('lobato' in k.get('NOME-PARA-CITACAO').lower()):

                if(n_autores ==1):
                    primeiro = primeiro+1

                if(n_autores == len(r)):
                    supervisor = supervisor+1


        autores.append(s_autores)


       
df = pd.DataFrame()
df['titulo'] = titulo
df['registro'] = codigo_registro
df['ano'] = ano
df['autores'] = autores


print('total:', total, 'primeiro:', primeiro, 'supervisor:', supervisor)

```

```python
def detected_language(x):
  translator = Translator()
  try:
    return LANGUAGES[translator.detect(x).lang]
  except:
    print("not detected")
    return "not detected"

def translate(x):
  translator = Translator()
  try:
    translation = translator.translate(x, dest='en')
    return translation.text
  except:
    print("not translated")
    return "not translated"
```

```python
df['titulo_traduzido'] = df['titulo'].apply(lambda x: translate(x))
```

```python
df['idioma'] = df['titulo'].apply(lambda x: detected_language(x)) 
```

```python


for index, row in df.iterrows():

# S. Miner- ação de dados do Enade de 2016 a 2018: uma análise sobre o município de Araçatuba. RENOTE 19, 2 (2021). [DOI]
    if(row.idioma == 'english'):
        
        print('\item['+row.ano+']\n', row.autores, row.titulo, '('+row.ano+'),', ' Register Code: '+row.registro)
       
    else:
        
      
        print('\item['+row.ano+']\n', row.autores, row.titulo_traduzido+' (In: '+str(row.idioma).title()+')', '('+row.ano+'),',' Register Code: '+row.registro)
       
                
 
```
