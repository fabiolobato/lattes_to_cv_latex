```python
from bs4 import BeautifulSoup
from googletrans import Translator
import pandas as pd
from googletrans import Translator
from googletrans import LANGUAGES
```

```python
# Reading data from the xml file
with open('cv3.xml', 'r', encoding='utf-8') as f:
    data = f.read()
 
```

```python
translator = Translator()
```

```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('CAPITULO-DE-LIVRO-PUBLICADO')


total = 0
primeiro = 0
supervisor = 0

titulo = []
doi = []
ano = []
home_page = []
livro = []
autores = []
isbn =[]

for i in b_unique:
    total = total+1
    a = i.find("DADOS-BASICOS-DO-CAPITULO")
   
    b = i.find("DETALHAMENTO-DO-CAPITULO")
    

    
    titulo.append(a.get('TITULO-DO-CAPITULO-DO-LIVRO').title())
    doi.append(a.get('DOI'))
    ano.append(a.get("ANO"))
    home_page.append(a.get('HOME-PAGE-DO-TRABALHO'))
    
    livro.append(b.get("TITULO-DO-LIVRO").title())
    isbn.append(b.get('ISBN'))
    
    
    
    r = i.find_all('AUTORES')
        
    n_autores = 0
    s_autores = ''
    for k in r:
        s_autores = s_autores+ k.get('NOME-PARA-CITACAO').title()+'; '
        n_autores = n_autores+1
        if('lobato' in k.get('NOME-PARA-CITACAO').lower()):
           
            if(n_autores ==1):
                primeiro = primeiro+1

            if(n_autores == len(r)):
                supervisor = supervisor+1
            

    autores.append(s_autores)
print(total, primeiro, supervisor)                
df = pd.DataFrame()
df['titulo'] = titulo
df['doi'] = doi
df['ano'] = ano
df['home_page'] = home_page
df['livro'] = livro
df['autores'] = autores
df['isbn'] = isbn

df.head(13)

        
    

```

```python
def detected_language(x):
  translator = Translator()
  try:
    return LANGUAGES[translator.detect(x).lang]
  except:
    print("not detected")
    return "not detected"

def translate(x):
  translator = Translator()
  try:
    translation = translator.translate(x, dest='en')
    return translation.text
  except:
    print("not translated")
    return "not translated"
```

```python
df['titulo_traduzido'] = df['titulo'].apply(lambda x: translate(x))
```

```python
df['idioma'] = df['titulo'].apply(lambda x: detected_language(x)) 
```

```python
def fixing_url(x):
    y=x
    if(x!= ''):
        s =x[-1] 
        print(x)
        if("\\" in s):
            y =  x[0:-1]
    return y
df['doi'] = df['doi'].apply(lambda x: fixing_url(x))
df['home_page'] = df['home_page'].apply(lambda x: fixing_url(x))
```

```python
df = df.sort_values(by=['ano'], ascending=False)
```
