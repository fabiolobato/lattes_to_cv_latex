```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('OUTRA-PRODUCAO-TECNICA')


titulo = []
ano = []
natureza = []



natureza = []

for i in b_unique:

    a = i.find("DADOS-BASICOS-DE-OUTRA-PRODUCAO-TECNICA")
   
    b = i.find("DETALHAMENTO-DE-OUTRA-PRODUCAO-TECNICA")
    

    natureza.append(a.get('NATUREZA'))
    titulo.append(a.get('TITULO'))
    ano.append(a.get("ANO"))

              
df = pd.DataFrame()
df['titulo'] = titulo
df['natureza'] = natureza
df['ano'] = ano


        
    

```

```python
df = df.sort_values(by=['ano'], ascending=False)
```
