```python
from bs4 import BeautifulSoup
from googletrans import Translator
```

```python
# Reading data from the xml file
with open('cv3.xml', 'r', encoding='utf-8') as f:
    data = f.read()
 
```

```python
translator = Translator()
```

```python

bs_data = BeautifulSoup(data, 'xml')
 
b_unique = bs_data.find_all('OUTRAS-ORIENTACOES-CONCLUIDAS')

lista = []
for i in b_unique:
    a = i.find("DADOS-BASICOS-DE-OUTRAS-ORIENTACOES-CONCLUIDAS")
   
    b = i.find("DETALHAMENTO-DE-OUTRAS-ORIENTACOES-CONCLUIDAS")
    if(a.get("NATUREZA")== "INICIACAO_CIENTIFICA"):
        instituicao = b.get('NOME-DA-INSTITUICAO')
        natureza = a.get("TITULO")
        titulo = a.get('TITULO')
        curso = b.get('NOME-DO-CURSO')
        orientado = b.get('NOME-DO-ORIENTADO')
        tipo_orientacao = b.get('TIPO-DE-ORIENTACAO')
        ano = a.get("ANO")


        
        s= 'Supervisor'
        
       # print(ano, orientado, tipo_orientacao, curso, natureza, instituicao)
        #  Alessandro Aparecido Barcellos, Postgraduate Specialization in Computing Applied to Education, CAE-USP. 
        #Thesis title: Open Data Mining - ENEM 2018 (in pt-BR). Graduated: Dec 2020.
        tag = "\\item["+str(ano)+"]"+'\n'+"{\em "+orientado.title()+"}"+'\n'+translator.translate(curso, dest='en').text+", "+translator.translate(instituicao, dest='en').text+"."+" Work: "+translator.translate(titulo, dest='en').text.capitalize()+" (in pt-BR)."+ " Role: \\textit{"+s+"}.\n"

        lista.append(tag)
 
```

```python
lista.reverse()
for i in lista:
    print(i)
```
