```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('TRABALHO-EM-EVENTOS')


total = 0
primeiro = 0
supervisor = 0

titulo = []
doi = []
ano = []
classificacao = []
home_page = []
livro = []
autores = []

natureza = []

for i in b_unique:
    total = total+1
    a = i.find("DADOS-BASICOS-DO-TRABALHO")
   
    b = i.find("DETALHAMENTO-DO-TRABALHO")
    

    natureza.append(a.get('NATUREZA'))
    titulo.append(a.get('TITULO-DO-TRABALHO').title())
    doi.append(a.get('DOI'))
    ano.append(a.get("ANO-DO-TRABALHO"))
    home_page.append(a.get('HOME-PAGE-DO-TRABALHO'))

    classificacao.append(b.get("CLASSIFICACAO-DO-EVENTO"))
    livro.append(b.get("NOME-DO-EVENTO").title())




    r = i.find_all('AUTORES')

    n_autores = 0
    s_autores = ''
    for k in r:
        s_autores = s_autores+ k.get('NOME-PARA-CITACAO').title()+'; '
        n_autores = n_autores+1
        if('lobato' in k.get('NOME-PARA-CITACAO').lower()):

            if(n_autores ==1):
                primeiro = primeiro+1

            if(n_autores == len(r)):
                supervisor = supervisor+1


    autores.append(s_autores)
print(total, primeiro, supervisor)                
df = pd.DataFrame()
df['titulo'] = titulo
df['evento'] = livro
df['ano'] = ano
df['natureza'] = natureza
df['home_page'] = home_page

df['autores'] = autores
df['classificacao'] = classificacao
df['doi'] = doi
df.head(30)

        
    

```

```python
def detected_language(x):
  translator = Translator()
  try:
    return LANGUAGES[translator.detect(x).lang]
  except:
    print("not detected")
    return "not detected"

def translate(x):
  translator = Translator()
  try:
    translation = translator.translate(x, dest='en')
    return translation.text
  except:
    print("not translated")
    return "not translated"
```

```python
df['titulo_traduzido'] = df['titulo'].apply(lambda x: translate(x))
```

```python
df['idioma'] = df['titulo'].apply(lambda x: detected_language(x)) 
```

```python
def fixing_url(x):
    y=x
    if(x!= ''):
        s =x[-1] 
        print(x)
        if("\\" in s):
            y =  x[0:-1]
    return y
df['doi'] = df['doi'].apply(lambda x: fixing_url(x))
df['home_page'] = df['home_page'].apply(lambda x: fixing_url(x))
```
