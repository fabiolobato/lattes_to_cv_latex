```python
bs_data = BeautifulSoup(data, 'xml')
b_unique = bs_data.find_all('ORGANIZACAO-DE-EVENTO')

#Parte A
natureza = []
titulo = []
ano = []

#Parte B
promotora = []
local = []
cidade = []

autores = []


for i in b_unique:

    a = i.find("DADOS-BASICOS-DA-ORGANIZACAO-DE-EVENTO")
    
    natureza.append(a.get('NATUREZA'))
    titulo.append(a.get('TITULO'))
    ano.append(a.get('ANO'))
   
    b = i.find("DETALHAMENTO-DA-ORGANIZACAO-DE-EVENTO")
    
    promotora.append(b.get('INSTITUICAO-PROMOTORA'))
    local.append(b.get('LOCAL'))
    cidade.append(b.get('CIDADE'))
    




    r = i.find_all('AUTORES')

    s_autores = ''
    for k in r:
        s_autores = s_autores+ k.get('NOME-PARA-CITACAO').title()+'; '
     
    autores.append(s_autores)


       
df = pd.DataFrame()


df['natureza'] = natureza
df['titulo'] = titulo
df['ano'] = ano
df['promotora'] = promotora
df['local'] = local
df['cidade'] = cidade
df['autores'] = autores


```

```python
df = df.sort_values(by=['ano'], ascending=False)
```

```python
def detected_language(x):
  translator = Translator()
  try:
    return LANGUAGES[translator.detect(x).lang]
  except:
    print("not detected")
    return "not detected"

def translate(x):
  translator = Translator()
  try:
    translation = translator.translate(x, dest='en')
    return translation.text
  except:
    print("not translated")
    return "not translated"
```

```python
df['titulo_traduzido'] = df['titulo'].apply(lambda x: translate(x))
```

```python
df['idioma'] = df['titulo'].apply(lambda x: detected_language(x)) 
```
