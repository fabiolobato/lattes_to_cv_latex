```python
with open('untitled.txt', 'r', encoding='UTF-8') as f:
    data = f.readlines()
    
```

```python
count = 0
for i in data:
    print(i)
    count = count+1
    if(count >=20):
        break;
```

```python
print('Period', '&', 'Level', '&', 'Program', '&', 'Courses', '\\\\')
for i in data:

    if '\item' in i:
        inicio = i.split('[')[1].split('--')[0]
        fim = i.split('[')[1].split('--')[1].split(']')[0]
       
    if 'Level:' in i:
        level = i.split(':} ')[1].strip()
    if 'Program:' in i:
        program =i.split(':} ')[1].strip()
    if 'Courses' in i:
        courses =i.split(':}')[1].strip().split(';  ')

        if(len(courses)==1):
            print(inicio, '--', fim, '&', level, '&', program, '&', courses[0], '\\\\')
        else:
            for j in range(len(courses)):
                if(j==0):
                    print(inicio, '--', fim, '&', level, '&', program, '&', courses[0], '\\\\')
                else:
                    print( '&', '&','&', courses[j], '\\\\')
                    
                    
           
        
    
    
```

```python
print('Period', '&', 'Courses', '\\\\')
previous =[]
for i in data:
    

    if '\item' in i:
        inicio = i.split('[')[1].split('--')[0]
        fim = i.split('[')[1].split('--')[1].split(']')[0]
        periodo = inicio+ '--'+ fim
        previous.append(periodo)

            
           
    if 'Level:' in i:
        level = i.split(':} ')[1].strip()
    if 'Program:' in i:
        program =i.split(':} ')[1].strip()
    if 'Courses' in i:
        courses =i.split(':}')[1].strip().split('; ')
        

        if(len(courses)==1):
            if(len(previous)<=1):
                print(inicio, '--', fim,  '&', courses[0], '\\\\')
            else:
                if(previous[-1] != previous[-2]):
                    print(inicio, '--', fim,  '&', courses[0], '\\\\')
                else:
                    print( '&', courses[0], '\\\\')
                    
        else:
            for j in range(len(courses)):
                if(j==0):
                    print(inicio, '--', fim,   '&', courses[0], '\\\\')
                else:
                    print( '&',  courses[j], '\\\\')
                    
                    
           
        
    
    
```
